```python
import pandas
import random
import hashlib
```

```python
insurance = pandas.read_csv('insurance.csv')
insurance.head
```

# Salting text with random integer

```python
def salt(input_text):
    return input_text+str(random.randint(1,256))
```

# Hashing function with sha256

```python
tmp_dataframe = pandas.DataFrame()
for index, row in insurance.iterrows():
    tmp_dataframe.loc[index,'hashed_bmi'] = hashlib.sha256(salt(str(row['bmi'])).encode('utf-8')).hexdigest()
insurance = pandas.concat([insurance, tmp_dataframe], axis=1, sort=False)
insurance.drop('bmi',axis=1,inplace=True)
```

# Hashing function with sha3 

```python
tmp_dataframe = pandas.DataFrame()
for index, row in insurance.iterrows():
    tmp_dataframe.loc[index,'hashed_charges'] = hashlib.sha3_512(salt(str(row['charges'])).encode('utf-8')).hexdigest()
insurance = pandas.concat([insurance, tmp_dataframe], axis=1, sort=False)
insurance.drop('charges',axis=1,inplace=True)
```

```python
insurance.to_csv('hashed_insurance.csv')
```
